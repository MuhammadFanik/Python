# Writing to a file - Text Files

### Case 1 - If  the file is not present

```python
# Open the file
f = open("sample.txt", "w")
# Write to the sample.txt file
f.write("This is a sample file")
# Close the file
f.close()

# This won't work, because the file is closed
# f.write("Writing after file is closed")
```

#### Write multiple Strings

```python
# Open the file
f = open("sample1.txt", "w")
# Writing multiple lines
f.write("Line 1")
f.write("\nLine 2")
# Closing the file
f.close()
```

### Case 2 - The file is already present

```python
# We want to write in sample.txt - If we write on an existing file, then the previous content is deleted and the new content overwrites it
f = open("sample.txt", "w")
f.write("Salman Khan")
f.close()
```

#### Problem with w mode - introducing the append mode

```python
# Adding more lines without removing the previous content
f = open("sample.txt", "a")
f.write("\nArijit Singh")
f.write("\nVirat Kohli")
f.close()
```

#### Write multiple lines

```python
# Each member of the list represents a line
l = ["l1", "\nl2", "\nl3", "\nl4", "\nl5", "\nl6" ]

f = open("sample.txt", "a")
f.writelines(l)
f.close()
```

## Reading from files

```python
# read() --> Aik saath poora content read karega
# readline() --> Reads one line at a time

f = open("sample.txt", "r")
# f.read() reads the entire file as a single string nad stores it in "s"
s = f.read()
print(s)    # Prints the content of the memory
f.close()
```

```python
# Reading upto n characters
f = open("sample.txt", "r")
s = f.read(10)  # First 10 characters
print(s)
f.close()
```

```python
# readline() -- reads just one line at a time including the newline character
f = open("sample.txt", "r")
print(f.readline(), end="")
print(f.readline(), end="")
f.close()
```

## Using context manager (with)

```python
# with

# Opens the file in append mode and assigns it to f. THe key benefit is that the file is closed automatically when the with block ends
with open("sample1.txt", "a") as f:
    f.write("\nArijit Singh")
```

```python
# try f.read() now

# Opens the file in read mode. If the file does not exist , Python will raise a FileNotFoundError
with open("sample1.txt", "r") as f:
    print(f.read())     # f.read() reads the entire file content as a single string and passes it directly to print
```

```python
# Using read with specific number of characters - Can be used to read big files
with open('sample.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
```

# Reading big file in chunks

```python
# This big list simulates a big file
big_list = ["Arijit Singh\n" for i in range(10000)]

# Creating a big file with the content of big_list
with open("big.txt", "w") as f:
    f.writelines(big_list)
```

```python
with open("big.txt", "r") as f:
    chunk_size = 10    # 500 characters
    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size), end="----")
        f.read(chunk_size)
```
