# Exceptions

```python
# The try block will generate an exception because x is not defined. Since the try block raises an error, the except block will be executed
try:
    print(x)
except:
    print("An exception occurred")
```

### Many exceptions

```python
# you can define as many exception blocks as you want
try:
    print(k)
except NameError:
    print("variable k is not defined")
except:
    print("Something else went wrong")
```

### Else -- Defines a block of code to be executed if no errors were raised

```python
try:
    print("Hello")
except:
    print("Something else went wrong")
else:
    print("Nothing went wrong")
```

### Finally --> It will be executed regardless if the try block raises an error or not

```python
try:
    print(b)
except:
    print("Something went wrong")
finally:
    print("This line will run doesn't matter if there is any error or not")
```

### Basic try and Except

```python
try:
    num = int(input("Enter a number: "))    # What if the user enters "abc"
    result = 10 / num                       # What if the user divides by zero
    print(result)
except:
    print("Some error occurred")

# The try block tries to run the code
# the except block runs if anything goes wrong
```

### Catching Specific Exceptions

```python
# It's better to catch specific errors so you know what exactly went wrong

try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print(result)
except ZeroDivisionError:   # Triggered when dividing by zero
    print("Division by zero not possible")
except ValueError:
    print("Input should be a number")
```

### The else and finally blocks

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:   # Triggered when dividing by zero
    print("Division by zero not possible")
except ValueError:      # if user enters anything else than a number
    print("Input should be a number")
else:
    print(result)
finally:
    print("Program finished")
```

### Getting the error message

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:      # e holds the actual error object
    print(f"Error caught: {e}")
```

### Catching multiple exceptions at once

```python
try:
    l = [1, 2, 3]
    print(l[10])        # Index Error
    print(int("abc"))   # Value Error
except (IndexError, ValueError) as e:
    print(f"Either index or value error occurred: {e}")
```

## Else and Finally

```python
# Q1. Ask the user to enter a number and calculate its square root (you can use ** 0.5). Use else to print the result only on success, and finally to always print "Calculation attempted." Catch a ValueError if input is invalid.
try:
    num = float(input("Enter a number: "))
    if num < 0:
        raise ValueError("Negative numbers are not allowed")
    sqrt = round(num ** 0.5, 2)
except ValueError:
    print("Input should be a number")
else:
    print(f"The square root of {num} is {sqrt}")
finally:
    print("Calculation attempted")
```
