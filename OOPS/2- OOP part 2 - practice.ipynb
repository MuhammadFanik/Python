```python
# Create a Circle class with a radius attribute. Write a method area() and a method perimeter() that both use self.radius to calculate and print the result.

class Circle:
    # Radius is in meters
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        area = math.pi * (self.radius ** 2)
        return round(area, 2)

    def perimeter(self):
        perimeter = 2 * math.pi * self.radius
        return round(perimeter, 2)
```

```python
circle1 = Circle(3)
print(circle1.area())
print(circle1.perimeter())
```

```python
# Create a Rectangle class with length and width. Write methods area() and is_square() that checks if length equals width.

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        area = self.length * self.width
        return area

    def is_square(self):
        if self.length == self.width:
            print("It's a square")
        else:
            print("It's a rectangle")
```

```python
r1 = Rectangle(10, 10)
r1.is_square()
```

```python
r2 = Rectangle(9, 17)
r2.is_square()
```

```python
# Create a Calculator class with methods add(), subtract(), and a menu() method that calls either based on user input.

class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        self.menu()

    def menu(self):
        user_input = int(input("""
        1. Press 1 to Add
        2. Press 2 to subtract
        3. Press 3 to multiply
        4. Press 4 to divide
        5. Anything else to exit
        """))
        if user_input == 1:
            self.add()
        elif user_input == 2:
            self.subtract()
        elif user_input == 3:
            self.multiply()
        elif user_input == 4:
            self.divide()
        else:
            pass


    def add(self):
        print(self.num1 + self.num2)
    def subtract(self):
        print(self.num1 - self.num2)
    def multiply(self):
        print(self.num1 * self.num2)
    def divide(self):
        print(self.num1 / self.num2)
```

### Reference Variables

```python
# Create a Dog class. Create one object dog1 and assign dog2 = dog1. Change dog2's name and show that dog1's name changes too, then explain why in a comment.

class Dog:
    def __init__(self):
        self.name = "Mars"
```

```python
# Creating the first object
dog1 = Dog()
# Printing the first dog's name
print(f"Dog1 name before changing: {dog1.name}")

# Assigning another variable (dog2) to dog1
dog2 = dog1
# Changing dog2's name
dog2.name = "Kit"
# Checking dog1's name
print(f"dog1 name after changing: {dog1.name}")

# Why its changing?
# dog1 is the reference variable for the Dog() object and initially it's name holds the value "Mars". When I write dog2 = dog1, it means that dog2 is also a reference variable that is pointing to the same address as dog1. Point to remember here is that the variables dog1 and dog2 are just reference variables, they don't hold the actual object. They just have the address where these objects are made in the memory. So any changes that are made to dog2 are also reflected into dog1 because they both point to the same box in the memory
```

### Pass by reference

```python
# Question 1
# Create a Student class with a grade attribute. Write a function outside the class called upgrade(student) that changes the grade to "A". Show that the original object is modified.

class Student:
    def __init__(self, grade):
        self.grade = grade

# s1 is the object
def upgrade(student):
    # This goes to the same address in the memory and changes it to "A" instead of "C"
    student.grade = "A"
    print(f"Grade after changing: {student.grade}")
```

```python
# A student object is created in the memory with "grade = C". "s1" is the reference variable pointing to that objects address in the memory
s1 = Student("C")
print("Grade before change:", s1.grade)
```

```python
# Calls the upgrade function and passes the address of the object "s1", not the copy of it. Inside the function, "student" is another reference variable pointing to the same address as s1.
# At this point, both s1 and student are pointing to the same location
upgrade(s1)
```

### Encapsulation, Private Variables, Getters and Setters

```python
# Q1. Create a Person class with private variables __name and __age. Write getters and setters for both, with a validation in the setter that age cannot be negative.

class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    # Return the private variable "name"
    def get_name(self):
        return self.__name

    # Return the age
    def get_age(self):
        return self.__age

    # Setter for name that updates the "name"
    def set_name(self, new_name):
        self.__name = new_name

    # Write a setter for __age that updates the age but with a validation — if the new age is negative, print "Age cannot be negative" and don't update it
    def set_age(self, new_age):
        if new_age < 0:
            print("Age cannot be negative")
        else:
            self.__age = new_age
```

```python
# Creating an object of the class Person
person1 = Person("Mansoor", 20)

# Trying to access the name and age variables
# print(person1.__age)      # Error

# To get the age, use the get_age method
print(person1.get_name())

# Use the setter to change the age
person1.set_age(27)
```

```python
# Checking the age after changing it through set_age method
person1.get_age()
```

```python
# Question 2: Create a BankAccount class with a private __balance. Write a getter to view balance and a setter that only allows balance to be increased, never decreased

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    # Get the balance
    def get_balance(self):
        return self.__balance

    # Set the balance
    def set_balance(self, new_balance):
        if new_balance > self.__balance:
            self.__balance = new_balance
        else:
            print("Balance can only be increased, not decreased")
```
