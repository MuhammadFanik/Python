# Abstraction

```python
from abc import ABC, abstractmethod
```

```python
# Parent Class
class Shape(ABC):   # Inheriting it from ABC makes it abstract
    @abstractmethod
    def area(self):     # Abstract method - no implementation
        pass

    @abstractmethod
    def perimeter(self):    # Abstract method - no implementation
        pass

# Explanation --> If you look at the area and perimeter methods, they both just have pass inside them. No actual calculation, no logic, nothing. The Shape class is just saying: "Hey, whoever inherits from me, must have an area() and a perimeter() method. But I am not going to tell you HOW to calculate it. That's your problem"
# That's the defining part of what to do without telling how to do
```

```python
# Now if you try to create an object of Shape, Python won't allow because Shape is just a blueprint
s = Shape()
```

### Child classes must implement all the abstract methods

```python
# Each child class calculates the area in its own way. Both are forced to have an area() method because Shape demanded it
# Child class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):     # Must implement this. Circle's own way of calculating the area
        return 3.14 * (self.radius ** 2)

    def perimeter(self):    # Must implement this
        return 2 * 3.14 * self.radius


# Child Class
class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):         # Must implement this. Rectangle's own way of calculating the area
        return self.l * self.w
    def perimeter(self):        # Must implement this
        return 2 * (self.l + self.w)
```

#### The actual meaningful output comes from the child classes:

```python
# Creating objects of the child classes
c1 = Circle(5)
print(c1.area())
print(c1.perimeter())
```

```python
rect1 = Rectangle(4, 5)
print(rect1.area())
print(rect1.perimeter())
```

## Abstraction example --> Payment System

##### In the real world, there are multiple ways to pay — credit card, PayPal, cryptocurrency etc. All of them must be able to process a payment and issue a refund, but each one does it in their own way.

```python
# Step 1 - Import the necessary modules
from abc import ABC, abstractmethod

# Step 2 - Create an abstract class Payment with two abstract methods: process_payment(amount) — no implementation, just pass refund(amount) — no implementation, just pass
# Parent Class
class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass

    @abstractmethod
    def refund(self, amount):
        pass


# Step 3 - Create a child class CreditCard that implements both methods: process_payment(amount) prints "Processing credit card payment of {amount}" refund(amount) prints "Refunding {amount} to credit card"
# Child Class
class CreditCard(Payment):
    def process_payment(self, amount):
        print(f"Processing Credit Card payment of {amount}")
    def refund(self, amount):
        print(f"Refunding amount {amount}")

# Step 4 - Create a child class PayPal that implements both methods: process_payment(amount) prints "Processing PayPal payment of {amount}" refund(amount) prints "Refunding {amount} to PayPal account"
class PayPal(Payment):
    def process_payment(self, amount):
        print(f"Processing PayPal Payment of {amount}")
    def refund(self, amount):
        print(f"Refunding PayPal amount {amount}")
```

```python
# Step 6 -  Creating objects
cc = CreditCard()
pp = PayPal()
```

```python
cc.process_payment(100)
cc.refund(500)
```

```python
# Step 7: Store all payment objects in a list and loop through them:
payments = [cc, pp]
for payment in payments:
    payment.process_payment(100)
    payment.refund(500)
```
