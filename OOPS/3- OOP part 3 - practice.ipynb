#### # Q1 - Library and Book System. Library Class owns the Book Class

```python
class Library:
    def __init__(self, name, book):
        self.name = name    # Simple name of the library
        self.book = book    # book object

    def show_book_details(self):
        print(f"""
        Library: {self.name}
        Book Title: {self.book.title}
        Author: {self.book.author}
        Pages: {self.book.pages}
        """)

# This is just a simple class which will have the details of the Book
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
```

```python
# Creating a book object independently
book1 = Book("11 Rules of Life", "Chetan Bhagat", 120)
# Creating a Library Object
lib1 = Library("Pearson's Library", book1)
```

```python
lib1.show_book_details()
```

#### Q3- Order and Product System

```python
# Step 1:
class Product:
    def __init__(self, name, price, category):
        self.name = name            # name of the product (string)
        self.price = price          # price of the product (int)
        self.category = category    # category of the product (string)

# Step 2:
class Order:
    def __init__(self, order_id, product, quantity):
        self.order_id = order_id        # order id (int)
        self.quantity = quantity        # order quantity
        self.product = product          # product object

    # Step 3: Show order summary
    def show_order_summary(self):
        print(f"""
        Order ID: {self.order_id}
        Product Name: {self.product.name}
        Category: {self.product.category}
        Price per Unit: {self.product.price}
        Quantity: {self.quantity}
        Total Price: {self.quantity * self.product.price}
        """)
```

```python
# Creating a product object - This exists independently
prod1 = Product("Lays", 1.5, "Snacks")
# Creating an order product - The Product object is passed in from outside
ord1 = Order(101, prod1, 4)
# Printing the details
ord1.show_order_summary()
```

# Types of Inheritance

#### Single Inheritance - Bank System - Create a BankAccount parent class with account_holder and balance. Create a SavingsAccount child class that inherits from it with a unique method add_interest() that increases the balance by 10%.

```python
class BankAccount:
    # Constructor
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder        # Account holder name
        self.__balance = balance                    # Account balance

    # Getter method - Because balance is a private attribute
    def get_balance(self):
        return self.__balance

    def set_balance(self, new_balance):
        self.__balance = new_balance

# Child class
class SavingsAccount(BankAccount):
    def add_interest(self, interest_rate):
        balance = self.get_balance()
        updated_balance = balance + ((interest_rate/100) * balance)
        self.set_balance(updated_balance)
        return f"Amount after interest is applied is: {updated_balance}"
```

```python
sa = SavingsAccount("Simon", 10000)
print(sa.add_interest(20))
```

#### Task 3: Create a BankAccount parent class with account_holder and balance. Create a SavingsAccount child class with its own __init__ that adds interest_rate and uses super().__init__(). Write a method apply_interest() that increases balance by the interest rate.

```python
class BankAccount:
    def __init__(self, acc_holder, balance):
        self.acc_holder = acc_holder
        self.__balance = balance        # private variable

    # getter method for balance
    def get_balance(self):
        return self.__balance

    # Setter method for balance
    def set_balance(self, new_balance):
        self.__balance = new_balance


class SavingsAccount(BankAccount):
    def __init__(self, acc_holder, balance, interest_rate):
        super().__init__(acc_holder, balance)
        self.interest_rate = interest_rate
        updated_balance = self.get_balance() + ((interest_rate / 100) * self.get_balance())
        print("Updated balance:", updated_balance)
        self.set_balance(updated_balance)
```

```python
savings_acc = SavingsAccount("Sam", 10000, 10)
```

#### Task 2:
Create three classes PDFReport, ExcelReport, and WordReport each with a generate() method. Write a function generate_report(report) that calls report.generate(). Pass all three objects to this function one by one.

```python
class PDFReport:
    def generate(self):
        print("PDF Report")

class ExcelReport:
    def generate(self):
        print("Excel Report")

class WordReport:
    def generate(self):
        print("Word Report")
```

```python
# this is duck typing. Python doesnt care about the type of object you pass into a function. It only cares about whether that object has the method, python will call on it.
# In this example, python never checks if the report is a WordReport, ExcelReport, or a PDFReport. It simply asks one question, does this object have a generate() method?
def generate_report(report):
    report.generate()
```

```python
generate_report(PDFReport())
generate_report(ExcelReport())
generate_report(WordReport())
```

# Method Overriding

#### Task 1:
Create a Notification parent class with a send() method that prints "Sending a generic notification". Create three child classes EmailNotification, SMSNotification, and PushNotification that each override send() with their own specific message like "Sending email to {recipient}" etc.

```python
# Parent class
class Notification:
    def send(self):
        print("Sending a generic notification")

# Child class - Email
class Email(Notification):
    def send(self, recipient):
        print("Sending email to", recipient)

# Child class - SMS
class SMS(Notification):
    def send(self, recipient):
        print("Sending SMS to", recipient)

# Child Class - Push
class Push(Notification):
    def send(self, recipient):
        print("Sending Push Notification to", recipient)
```

```python
email = Email()
sms = SMS()
push = Push()
```

```python
email.send("Rizwan")
sms.send("Babar")
push.send("Shadab")
```

#### Task 2:
Create a Discount parent class with a calculate_discount(price) method that gives a 5% discount. Create three child classes SeasonalDiscount, LoyaltyDiscount, and StudentDiscount that each override calculate_discount() with their own discount percentage.

```python
class Discount:
    def __init__(self, price):
        self.price = price

    def calculate_discount(self):
        updated_price = self.price - (0.05 * self.price)
        print(f"Discounted Price: {updated_price}")

class SeasonalDiscount(Discount):
    def calculate_discount(self, discount_rate):
        updated_price = self.price - ((discount_rate / 100) * self.price)
        return f"Seasonal Price is: {updated_price}"

class LoyaltyDiscount(Discount):
    def calculate_discount(self, discount_rate):
        updated_price = self.price - ((discount_rate / 100) * self.price)
        return f"Loyalty Price is: {updated_price}"

class StudentDiscount(Discount):
    def calculate_discount(self, discount_rate):
        updated_price = self.price - ((discount_rate / 100) * self.price)
        return f"Student Price is: {updated_price}"
```

```python
sd = SeasonalDiscount(100)      # Price -- 100
ld = LoyaltyDiscount(100)       # Price -- 100
std = StudentDiscount(100)      # Price -- 100
```

```python
sd.calculate_discount(5)
```

```python
ld.calculate_discount(10)
```

```python
std.calculate_discount(20)
```

# Method Overloading

#### Task 1:
Create a Message class with a send(recipient, message, cc=None) method. If cc is provided print "Sending message to {recipient} and CC to {cc}", otherwise just print "Sending message to {recipient}".

```python
class Message:
    def send(self, recipient, message, cc = None):
        if cc != None:
            print(f"Sending message to {recipient} and CC to {cc}")
        else:
            print(f"Sending message to {recipient}")
```

```python
msg = Message()
```

```python
msg.send("Sam", "Hi")
```
