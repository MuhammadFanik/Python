# Aggregation

```python
class Customer:
    # I am passing an object (address1) to the constructor of a class (Customer). This is known as aggregation
    def __init__(self, name, gender, address):
        self.name = name            # self.name = "Arsalan"
        self.gender = gender        # self.gender = "Male"
        self.address = address      # self.address = address1

    # Because self.address is pointing to the same location where all the address details are stored, so I can access all the data available in there. In this method I am just printing out each variable, That's it
    # For the private__city variable: Since I have the Address object, I can call any public method that belongs to it, including get_city()
    def print_address(self):
        print(self.address.house_no, self.address.street_no, self.address.get_city(), self.address.pin_code, self.address.state)



# Just a simple class that holds the Address. Nothing new here
class Address:
    def __init__(self, house_no, street_no, city, pin_code, state):
        self.house_no = house_no        # self.house_noo = 7
        self.street_no = street_no      # self.street_no = 7
        self.__city = city                # self.city = "London"
        self.pin_code = pin_code        # self.pin_code = 15000
        self.state = state              # self.state = "Greater London"

    # Since city is now a private variable, I cannot access it directly from outside the class
    def get_city(self):
        return self.__city
```

```python
# Creating an Address object. A real Address object is created somewhere in the memory. "address1" is a reference variable that holds the memory address to that object(Address). This object exists independently (no customer involved yet)
address1 = Address(7, 7, "London", 15000, "Greater London")


# When I create the object for Customer class and I pass in the values, "Arsalan", "Male", and address1. Arsalan will go in self.name, Male will go in self.gender, but in the case of self.address it will be self.address = address1. This means that self.address is pointing to the same box in the memory where address1 is pointing to. As of now, the information about the address (which is 7, 7, London...), there are two reference variables pointing to this info. one is address1 and the other is self.address
customer1 = Customer("Arsalan", "Male", address1)
```

```python
address1.get_city()
```

```python
# Independence is the feature of aggregation. "address1" is created first independently, then passed onto customer1
```

## Method overriding

```python
class Phone:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def buy(self):
        print(f"I bought a {self.make} {self.model} {self.year} Mobile Phone")

class SmartPhone(Phone):
    def buy(self):
        print(f"I bought a {self.make} {self.model} {self.year} Smart Phone")
```

## Super keyword

### Super Basics

```python
# Parent class
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a Phone")


# Child Class
class SmartPhone(Phone):
    def buy(self):
        print("Buying a Smart Phone")
        # Syntax to call the super keyword
        super().buy()

# when I create an object "s", the values are initialized in the Phone constructor. After that when I am writing s.buy(), it calls the buy method in the child class because the child class has its own buy method. Now inside the buy method of the child class, it first prints a statement "Buying a smartphone" and after that its calling the buy method of the parent class using super keyword.
```

```python
s = SmartPhone(20000, "Samsung", 48)
s.buy()
```

### Another example on super

```python
# Parent class
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

# Child Class
class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print("Inside SmartPhone Constructor")
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
```

```python
s = SmartPhone(20000, "Samsung", 48, "Android", 6)
```

```python
print(s.price)
print(s.camera)
print(s.os)
print(s.ram)
```

```python
# Q1
class Parent:
    def __init__(self, num):
        self.__num = num    # self.__num=100

    # Getter method
    def get_num(self):
        return self.__num

class Child(Parent):
    def __init__(self, num, val):
        super().__init__(num)
        self.__val = val    # self.__val = 200

    def get_val(self):
        return self.__val
```

```python
son = Child(100, 200)
print(son.get_num())    # 100
print(son.get_val())    # 200
```

# Polymorphism

```python
# Basic Code from Claude - Think about the word "speak". If a dog speaks it barks, if a cat speaks it meows, if a human speaks it talks. The action is the same — speak — but the behavior is different depending on who is doing it. That's polymorphism.

# All three classes have same method name speak() but each behave differently.
class Dog:
    def speak(self):
        print("Bark")

class Cat:
    def speak(self):
        print("Meow")

class Human:
    def speak(self):
        print("Talk")

# This is an example of duck typing - refer to notes
```

```python
# Now the power of polymorphism
animals = [Dog(), Cat(), Human()]
for i in animals:
    i.speak()
```

## Method overloading

#### The main problem -- The second add() method completely overwrites the first one. Python only keeps the last definition

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def add(self, a, b, c):
        return a + b + c

calc1 = Calculator()
calc1.add(1, 2)
```

#### How to solve this problem?

##### Sol1: Default params

```python
class Calculator:
    def add(self, a, b, c=0):   # c is optional in this case. If user doesn't enter anything, its 0
        return a + b + c
```

```python
calc1 = Calculator()
# Passing in two params
print(calc1.add(1, 2))
# Passing in three params
print(calc1.add(1, 2, 3))
```

##### Sol2: *args - When you dont know how many params will be passed

```python
class Calculator:
    def add(self, *args):
        return sum(args)
```

```python
calc1 = Calculator()
```

```python
print(calc1.add(1)) # Passing in just one argument
print(calc1.add(1, 2, 4, 5))    # Passing in 4 arguments
```
