```python
## Write OOP classes to handle the following scenarios

##### A user can create and view 2D coordinates
##### A user can find out the distance between 2 2D coordinates
##### A user can find the distance of a coordinate from the origin
##### A user can check if a point lies on a given line
##### A user can find the shortest distance between a line and a 2D coordinate
```

```python
class Point:
    # Creates a point with x and y coordinates
    def __init__(self, x, y):
        self.x_coordinate = x
        self.y_coordinate = y

    # Step 1: A user can view 2D coordinates
    def __str__(self):  # Python automatically calls it when the object is printed
        return f"({self.x_coordinate}, {self.y_coordinate})"

    # Step 2: A user can find the distance between 2 coordinates
    # This function takes 2 parameters (self, other), and both are Point Objects. Self = point1, other = point2.
    # A method can receive another object of the same class as a parameter
    def euclidean_distance(self, other):
        return round((((self.x_coordinate - other.x_coordinate) ** 2) + ((self.y_coordinate - other.y_coordinate) ** 2)) ** 0.5, 3)

    # A user can find the distance from the origin. Instead of re-writing the distance formula, it creates a temporary Point(0, 0)
    # object on the spot and passes it into euclidean_distance()
    def origin_distance(self):
        return self.euclidean_distance(Point(0,0))


# A user can check if a point lies on the given line
class Line:
    # Line: ax + by + c = 0
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def __str__(self):
        return f"{self.a}x + {self.b}y + {self.c} = 0"

    # Self --> Line object
    # point --> Point object
    def point_on_line(self, point):
        # Here I am accessing the Point Class attributes (point.x_coordinate, point.y_coordinate) --> One object uses another objects attributes
        if self.a*point.x_coordinate + self.b*point.y_coordinate + self.c == 0:
            print(f"The point lies on the line")
        else:
            print("The point does not lie on the line")

    def shortest_distance(self,point):
        return abs(self.a*point.x_coordinate + self.b*point.y_coordinate + self.c)/(self.a**2 + self.b**2)**0.5
```

```python
point1 = Point(3, 4)
print(point1)
```

```python
# Step 2: A user can find the distance between 2 points
point2 = Point(7, 9)
# Self(Point1) --> the object calling the method
# Other(Point2) --> the object being passed in
# a method can receive another object of the same class as a parameter. In this case its point2
point1.euclidean_distance(point2)
```

```python
line1 = Line(7, 2, 4)
print(line1)
```

## How objects access attributes and methods

```python
class Person:
    def __init__(self, name, country):
        self.name = name
        self.country = country

    def greet(self):
        if self.country == "India":
            print(f"{self.name} is from India")
        else:
            print(f"{self.name} is from {self.country}")
```

```python
# How to access the attributes
person1 = Person("Arsalan", "India")
print(person1.name)
print(person1.country)
```

```python
# How to access the methods
person1.greet()
```

## Reference Variables

#### Object without a reference

```python
class Person:
    def __init__(self):
        self.name = "Arsalan"
        self.age = 35
```

```python
# Actually, p is not the object. When we call Person(), object is created at that moment, and we store the reference/address of that object in the variable "p"
p = Person()
```

```python
# q and p, both are variables pointing to the same object
q = p

# Multiple Referencing
print(id(p))
print(id(q))
```

## Pass by reference

```python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender


# Outside
def greet(p1):
    print(f"Hi, my name is {p1.name} and I am a {p1.gender}")
```

```python
p1 = Person("Arsalan", "Male")
```

```python
greet(p1)
```
